```python
import json
import pandas as pd
from pandas.io.json import json_normalize
import numpy as np
from tqdm import tqdm_notebook as tqdm
```

```python
# Import songs from dataset
# csv_songs = pandas.read_csv()
columns = ['artist_name', 'title']
million_csv = pd.read_csv('MillionSongDB.csv', skipinitialspace=True, usecols=columns)
# artist_names = ['claris', 'asdf', 'claris']
# track_names = ['colorful', 'asdf', 'step']
million_csv['spotify_id'] = 'n/a'
print(million_csv.iloc[[0]]['artist_name'][0])
print(million_csv.iloc[[0]]['title'][0])
```

```python
artist_list = million_csv['artist_name']
song_list = million_csv['title']
```

```python
songs_new = []
with open('tried_to_find.txt') as json_file:
     songs_new = json.load(json_file)
```

```python
songs_old = []
with open('found_song_ids_json.txt') as json_file:
    songs_old = json.load(json_file)
```

```python
print(songs_new[0])
songs_old = json.loads(songs_old)
print(songs_old[0])
```

```python
def extract_old(song_old):
    spotify_id = song_old['spotify_id']
    artist_name = song_old['artist_name']
    title = song_old['title']
    return spotify_id, artist_name, title

def extract_new(song_new):
    artist_name = song_new[0]
    title = song_new[1]
    spotify_id = song_new[2]
    return spotify_id, artist_name, title
    
    
song_ids = []
for index in range(0, len(million_csv)):
    looking_song = song_list[index]
    looking_artist = artist_list[index]
    
    found = False
    for element in songs_old:
        if(found):
            break
        spotify_id, artist_name, title = extract_old(element)
        if(looking_song == title and looking_artist == artist_name):
            song_ids.append(spotify_id)
            print("Matched ", index)
            found = True

    for element in songs_new:
        if(found):
            break
        spotify_id, artist_name, title = extract_new(element)
        if(looking_song == title and looking_artist == artist_name):
            song_ids.append(spotify_id)
            print("Matched ", index)
            found = True
        
    if(not found):
        print("Lost ", index)
        song_ids.append(None)
```

```python
print(len(song_ids))
```

```python
d = {'artist_name': artist_list, 'title':song_list, 'spotify_id':song_ids }
df = pd.DataFrame(data=d)
```

```python
df.to_json(r'full_found_ids_json.txt')
```

```python
import spotipy
import json
import spotipy.oauth2 as oauth2
import pandas as pd
from pandas.io.json import json_normalize
import numpy as np
import time
from tqdm import tqdm_notebook as tqdm
```

```python
df = pd.read_json('full_found_ids_json.txt')
df.sort_index()
```

```python
# Spotify Credentials Creation
credentials = oauth2.SpotifyClientCredentials(
        client_id='',
        client_secret='')

token = credentials.get_access_token()
sp = spotipy.Spotify(auth=token)
```

```python
# filler song
results = sp.search(q='artist:'+str('claris')+' track:'+str('shiori'), limit=1)
track_id = results['tracks']['items'][0]['id']
filler_song_features = sp.audio_features(track_id)
for item in filler_song_features[0]:
    filler_song_features[0][item] = np.nan
print(filler_song_features[0])
```

```python
features = []
with open('spotify_features.txt') as json_file:
    features = json.load(json_file)
len(features)
```

```python
# features = []
# Grab the audio features in sets
for index in tqdm(range(len(features), len(df))):
    spot_id = df.iloc[index].spotify_id
    if(spot_id == 'None'):
        features.append(filler_song_features)
    else:
        counter = 0
        while True:
            counter = counter + 1
            try:
                features.append(sp.audio_features(spot_id))
            except:
                time.sleep(1)
                if(counter == 5):
                    features.append(filler_song_features)
                    print("Failed too many times")
                    break
                continue
            break
        time.sleep(.25)
```

```python
with open('spotify_features.txt', 'w') as outfile:
    json.dump(features, outfile)
```
