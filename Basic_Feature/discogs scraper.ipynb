```python
import pandas as pd
import time
from tqdm import tqdm_notebook as tqdm
import discogs_client
```

```python
million_csv = pd.read_csv('MusicRerunDBMB.csv', skipinitialspace=True)
```

```python
# print(million_csv)
zero_years = million_csv[['Unnamed: 0','title', 'artist_name', 'year']]
zero_years = zero_years.loc[zero_years['year'] == '0']
zero_years = pd.read_csv('unfound_years2.csv')
print(zero_years)
```

```python
def remove_dumb(song_name):
    fixed_song = str(song_name)
    index = fixed_song.find('(')
    if(index != -1):
        fixed_song = fixed_song[0:index]
    fixed_song = fixed_song.replace('(', '')
    fixed_song = fixed_song.replace(')', '')
    fixed_song = fixed_song.replace('Version ', '')
    fixed_song = fixed_song.replace('Album ', '')
    fixed_song = fixed_song.replace('LP ', '')
    fixed_song = fixed_song.replace('featuring ', '')
    fixed_song = fixed_song.replace('mix', '')
    fixed_song = fixed_song.replace('and ', '')
    fixed_song = fixed_song.replace('And ', '')
    fixed_song = fixed_song.replace('& ', '')
    return fixed_song
```

```python
d = discogs_client.Client('YearFinder/1.0', user_token ="")
```

```python
def find_year(artist_name, title):
    results = d.search(artist_name + ' ' + title, type='release')
    try:
        year = results[0].year
#         release = d.release(id)
        print(year)
        return year
    except:
        print(title, 'by', artist_name, "not found")
        return 0
```

```python
for index, row in tqdm(zero_years.iterrows()):
    if(row['year'] != 0 and row['year'] != '0'):
        continue
    zero_years.loc[index, 'year'] = find_year(remove_dumb(row['artist_name']), remove_dumb(row['title']))
    time.sleep(1)
```

```python
zero_years.to_csv ('unfound_years2.csv', index = None, header=True)
```
