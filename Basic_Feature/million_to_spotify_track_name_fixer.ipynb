```python
import spotipy
import json
import spotipy.oauth2 as oauth2
import pandas as pd
from pandas.io.json import json_normalize
import numpy as np
import time
from tqdm import tqdm_notebook as tqdm
```

```python
songs = []
with open('tried_to_find.txt') as json_file:
     songs = json.load(json_file)
```

```python

credentials = oauth2.SpotifyClientCredentials(
        client_id='',
        client_secret='')

token = credentials.get_access_token()
sp = spotipy.Spotify(auth=token)
```

```python
failed_songs = []
def find_song_id(artist_name, track_name):
#     print(artist_name)
#     print(track_name)
    results = sp.search(q='artist:'+str(artist_name)+' track:'+str(track_name), limit=1)
    if(results['tracks']['total'] == 0):
        failed_songs.append(str(str(artist_name) +";;;" + str(track_name)))
        return None
    else:
        track_id =  results['tracks']['items'][0]['id']
        return track_id
```

```python
# song = split_content[0]
def remove_dumb(song_name):
    fixed_song = str(song_name)
    index = fixed_song.find('(')
    if(index != -1):
        fixed_song = fixed_song[0:index]
    fixed_song = fixed_song.replace('(', '')
    fixed_song = fixed_song.replace(')', '')
    fixed_song = fixed_song.replace('Version ', '')
    fixed_song = fixed_song.replace('Album ', '')
    fixed_song = fixed_song.replace('LP ', '')
    fixed_song = fixed_song.replace('featuring ', '')
    fixed_song = fixed_song.replace('mix', '')
    fixed_song = fixed_song.replace('and ', '')
    fixed_song = fixed_song.replace('And ', '')
    fixed_song = fixed_song.replace('& ', '')
    return fixed_song


tids = []
song_missing = []
filler_song = find_song_id('claris', 'shiori')

for index, song in tqdm(enumerate(songs)):
    song_name = song[1]
    artist_name = song[0]
    found = song[2]
#     print(found, song_name, artist_name)
    if(found is not None):
        continue
    song_fixed = remove_dumb(song_name)
#     print(song_fixed)
#     time.sleep(.5)
    try:
        track_id = find_song_id(artist_name, song_fixed)  
    except:
        track_id = None
        print("Failed to request ID")
    songs[index][2] = track_id
    if(songs[index][2] is not None):
        print(songs[index][2])

    
```
