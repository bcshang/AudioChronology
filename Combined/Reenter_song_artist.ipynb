```python
import numpy as np
import pandas as pd
import math
```

```python
data_pred = pd.read_csv("ProbPred.csv")
```

```python
data = pd.read_csv("Complete_musicDB.csv")
rows = data.shape[0]
artist = []
song = []
Y = data['year']


for cur in range(0, rows):
    if(Y[cur] == 0 or math.isnan(Y[cur])):
        print(cur)
    else:
        #get index of the years that were taken out
        artist.append(data['artist_name'][cur])
        song.append(data['title'][cur])
print(artist)
print(song)
```

```python
artist_df = pd.DataFrame(artist, columns = ['artist'])
song_df = pd.DataFrame(song, columns = ['title'])
print(artist_df)
```

```python
print(song_df)
```

```python
final = artist_df.join(data_pred)
final = song_df.join(final)
print(final.columns.values)
```

```python
final["title_lower"] = final["title"].str.lower()
```

```python
final = final.drop('title', axis = 1)
print(final.columns.values)
```

```python
row = final.shape[0]
for cur in range(0, row):
    test_string = str(final['title_lower'][cur]).replace('/', '')
    test_string = test_string.replace('.', '')
    test_string = test_string.replace('?', '')
    test_string = test_string.replace('!', '')
    test_string = test_string.split('_')[0]
    final['title_lower'][cur] = test_string
```

```python
final_sorted = final.sort_values(by ='title_lower' )
```

```python
print(final_sorted.title_lower)
```

```python
final_sorted.to_csv('basic_artist_title.csv', index=False)
```
