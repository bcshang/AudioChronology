```python
import pandas as pd
import numpy as np
```

```python
basic = pd.read_csv("basic_index.csv")
freq = pd.read_csv("freq_domain_pred.csv")
spec = pd.read_csv("time_series_pred.csv")
```

```python
complete_data = pd.read_csv("Complete_musicDB.csv")
complete_Y = complete_data['year']
basic = basic.join(complete_Y)
print(basic)
```

```python
basic_combined_freq = pd.DataFrame()
for freq_idx in freq.idx:
    row = pd.DataFrame(basic.iloc[[freq_idx]])
    basic_combined_freq = basic_combined_freq.append(row)
```

```python
print(len(freq))
print(len(basic_combined_freq))
print(basic_combined_freq)
basic_combined_freq.to_csv("basic_filtered.csv", index=False)
```

```python
basic_fil = pd.read_csv("basic_filtered.csv")
basic_fil.columns = ['1910_basic', '1920_basic', '1930_basic', '1940_basic', '1950_basic', '1960_basic', '1970_basic', '1980_basic', '1990_basic', '2000_basic', '2010_basic', 'year']
combined_audio = combined_spec.join(basic_fil)
```
