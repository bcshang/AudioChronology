```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
random.seed(42)
```

```python
df = pd.read_csv("tf_idf.csv")
df = df.rename(columns={"Unnamed: 0":"song_file_name"})
df
```

```python
years = [file.split("_")[-1] for file in df['song_file_name']]
year_length = [len(year) for year in years]
df['year_length'] = year_length
```

```python
df_tfidf = df[df.year_length == 4]
```

```python
df_tfidf.head()
```

```python
years = [int(file.split("_")[-1]) for file in df_tfidf['song_file_name']]
decades = [year//10*10 for year in years]
```

```python
df_tfidf['decade'] = decades
df_tfidf = df_tfidf.drop('year_length', axis=1)
```

```python
df_tfidf.head()
```

```python
np.unique(df_tfidf['decade'].values)
```

```python
df_tfidf = df_tfidf[df_tfidf.decade != 1920]
```

```python
print(df_tfidf['decade'].value_counts())
ax = sns.countplot(df_tfidf['decade'])
```

```python
decade_labels = {
  1950: 0,
  1960: 1,
  1970: 2,
  1980: 3,
  1990: 4,
  2000: 5,
  2010: 6
}
labels = [decade_labels[x] for x in df_tfidf['decade']]
df_tfidf['label'] = labels
```

```python
df_tfidf = df_tfidf.set_index('song_file_name')
df_tfidf.head()
```

```python
df_tfidf = df_tfidf.drop('decade', axis=1)
df_tfidf.shape
```

```python
df_tfidf.head()
```

```python
df_tfidf.to_csv('tfidf_data_labeled.csv')
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn import metrics
import xgboost as xgb
from xgboost import XGBClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_tfidf.loc[:,:'confusi'], df_tfidf['label'],
                                                    test_size=0.20, random_state=42)
```

```python
model_01 = XGBClassifier(objective="multi:softmax", random_state=42)
model_01.fit(X_train, y_train)
```

```python
y_pred = model_01.predict(X_test)
y_pred
```

```python
print(metrics.classification_report(y_test, y_pred))
print(metrics.accuracy_score(y_pred, y_test))
```

```python
dtrain = xgb.DMatrix(data=X_train, label=y_train)
dtest = xgb.DMatrix(data=X_test)
```

```python
params = {
    'max_depth': 3,
    'objective': 'multi:softmax',
    'num_class': 7,
    'n_gpus': 0
}
```

```python
bst = xgb.train(params, dtrain)
```

```python
pred = bst.predict(dtest)
print(metrics.classification_report(y_test, pred))
```

```python
xgb1 = XGBClassifier(
    learning_rate=0.01, 
    n_estimators=1000, 
    max_depth=5,
    min_child_weight=1, 
    gamma=0,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="multi:softmax",
    nthread=4,
    scale_pos_weight=1,
    seed=42,
    num_class=7
    )
xgb1.fit(X_train, y_train)
```

```python
y_pred_01 = xgb1.predict(X_test)
print(metrics.classification_report(y_test, y_pred_01))
print(metrics.accuracy_score(y_pred_01, y_test))
```

```python
xgb_params_01 = {
    'max_depth':range(3,10),
    'min_child_weight':range(1,6)
}
estimator_01 = XGBClassifier(
    learning_rate=0.01, 
    n_estimators=1000,
    max_depth=5,
    min_child_weight=1,
    gamma=0,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="multi:softmax",
    nthread=4,
    scale_pos_weight=1,
    seed=42,
    num_class=7
    )
xgb_model_01 = GridSearchCV(estimator=estimator_01, 
                            param_grid=xgb_params_01, 
                            n_jobs=4, 
                            scoring='accuracy',
                            iid=False,
                            cv=5)
xgb_model_01.fit(X_train, y_train)
```

```python
print("Best params: ", xgb_model_01.best_params_)
print("Best score: ", xgb_model_01.best_score_)
```
