```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
random.seed(42)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
```

```python
np.random.seed(42)
```

```python
df = pd.read_csv("tf_idf.csv")
df = df.rename(columns={"Unnamed: 0":"song_file_name"})
df.head()
```

```python
years = [file.split("_")[-1] for file in df['song_file_name']]
year_length = [len(year) for year in years]
df['year_length'] = year_length
```

```python
df_tfidf = df[df.year_length == 4]
```

```python
df_tfidf.head()
```

```python
years = [int(file.split("_")[-1]) for file in df_tfidf['song_file_name']]
decades = [year//10*10 for year in years]
```

```python
df_tfidf['decade'] = decades
df_tfidf = df_tfidf.drop('year_length', axis=1)
```

```python
df_tfidf.head()
```

```python
print(df_tfidf['decade'].value_counts())
ax = sns.countplot(df_tfidf['decade'])
```

```python
df_tfidf = df_tfidf.set_index('song_file_name')
df_tfidf.head()
```

```python
tfidf_dataset = df_tfidf.values
X = tfidf_dataset[:,0:-1]
Y = tfidf_dataset[:,-1]
```

```python
encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
# convert integers to dummy variables (i.e. one hot encoded)
dummy_y = np_utils.to_categorical(encoded_Y)
```

```python
# define baseline model
def baseline_model():
    model = Sequential()
    model.add(Dense(8, input_dim=11078, activation='relu'))
    model.add(Dense(8, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
estimator = KerasClassifier(build_fn=baseline_model, epochs=200, batch_size=5, verbose=0)
```

```python
kfold = KFold(n_splits=10, shuffle=True, random_state=42)
```

```python
results = cross_val_score(estimator, X, dummy_y, cv=kfold)
print("Baseline: %.2f%% (%.2f%%)" % (results.mean()*100, results.std()*100))
```
